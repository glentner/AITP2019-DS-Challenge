```python
final = pd.read_csv("/scratch/brown/train52/data/obs_IN.csv")
```

```python
final.head()
```

```python
date = final.datetime.apply(str)
date = date.str.split('[/-]',expand=True)
date[0].unique()
```

```python
final['year']=date[0]
final['month']=date[1]
final['day']=date[2]
```

```python
final.head(10)
```

```python
data = final.drop(['call_name', 'country'], axis=1)
```

```python
data.head()
```

```python
data_temp = data[data['measure_id'] == 1.0]
```

```python
mean_temp = data_temp.groupby(['latitude', 'longitude'], as_index=False)['measure_value'].mean()
mean_temp.head(10)
```

```python
temp_regression_values = mean_temp.values
```

```python
X = temp_regression_values[:,0:2]
X.shape
```

```python
y = temp_regression_values[:,2]
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
print(y_pred)
# Plot outputs
```
