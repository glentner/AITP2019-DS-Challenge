```python
import csv
```

```python
IN_data = []
IN_stationIds = []

with open("/home/glentner/public/datasets/noaa/gsod.stn.csv", 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print "columns are: " + ", ".join(row)
            line_count += 1
        else:
            if row[4] == "IN":
                IN_data.append(row)
                if row[0] not in IN_stationIds:
                    IN_stationIds.append(row[0])
                
print len(IN_data)
print IN_stationIds
```

```python
station_data = []

with open("data/gsod.obs.csv", 'r') as csv_obs_file:
    count = 0
    multiplier = 0
    for line in csv_obs_file:
        stat_id = line.index(",")
        substr = line[:stat_id]
        if substr in IN_stationIds:
            station_data.append(line)
        count += 1
        if count > 10000000:
            percent = (multiplier * 1e7 + count) / 2291730018.0
            print '\r', len(station_data), percent
            count = 0
            multiplier += 1
        
print len(station_data)
```

---

```python
with open('data/us-in.csv', mode='w') as output:
    for line in station_data:
        if line[:6] != '999999':
            output.write(line)
```
