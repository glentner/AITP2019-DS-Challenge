```python
file ="/home/glentner/public/datasets/noaa/gsod.obs.csv"
import pandas as pd
```

```python
stationFile = "/home/glentner/public/datasets/noaa/gsod.stn.csv"
station_data = pd.read_csv(stationFile)
```

```python
in_data = station_data[(station_data["state"] == "IN")]
st_id = in_data["station_id"]
wb_id = in_data["wban_id"].tolist()
list1 = st_id.tolist()
ulist = list(set(list1))
uwb_id = list(set(wb_id))
uwb_id
```

```python
import csv
data = []
index = 0
with open(file, "r") as fptr:
    for line in fptr:
        if line.split(",")[0] in ulist:
            data.append(line)
    
```

```python
data
```

```python
filetowname = "./in.csv"
filetow = open(filetowname,'w')
for line in data:
    filetow.write(line)
```

```python
datan99 = []
for line in data:
    if line.split(",")[0] != '999999':
            datan99.append(line)
```
